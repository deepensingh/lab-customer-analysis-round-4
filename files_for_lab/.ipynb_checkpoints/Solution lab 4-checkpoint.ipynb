```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cfile = pd.read_csv('marketing_customer_analysis.csv') 
```

```python
cfile.head(10)
```

# Check the data types of the columns. Get the numeric data into dataframe called numerical and categorical columns in a dataframe called categoricals. 14 (You can use np.number and np.object to select the numerical data types and categorical data types respectively) 15

```python
numerical = pd.DataFrame(cfile.select_dtypes(np.number))
```

```python
numerical.head(5)
```

```python
categorical = pd.DataFrame(cfile.select_dtypes(object))
```

```python
categorical.head(5)
```

# Now we will try to check the normality of the numerical variables visually 16
* Use seaborn library to construct distribution plots for the numerical variables 17
* Use Matplotlib to construct histograms 18
* Do the distributions for different numerical variables look like a normal distribution 1

```python
for i in numerical.columns:
    print("Plotting columns", i)
    sns.displot(x = i, data = numerical)
    plt.show()
```

```python
for i in numerical.columns:
    print('plotting columns', i)
    plt.hist(x = cfile[i], density = 'true')
    plt.ylabel('count')
    plt.xlabel(i)
    plt.show()
```

```python
cfile.corr()
```

```python
correlation_matrix = cfile.corr()
plt.figure(figsize = (9,5))
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
